Import Library dan mempersiapkan Dataset

```python
import pandas as pd
df1 = pd.read_csv("decisiontree_ch6.csv")
print(df1)
print("----------")
encoding = {"mesin": {"bensin": 0, "diesel": 1},
            "penggerak": {"depan": 0, "belakang": 1}}
df1.replace(encoding, inplace=True)
print(df1)
```

Kolom "ID" tidak diperlukan dalam data training, sehingga perlu kita buang

```python
X = df1.drop(['ID','label'], axis=1)
y = df1['label']
```

Split dataset menjadi data training dan data test

```python
import sklearn.model_selection as ms
X_train, X_test, y_train, y_test = ms.train_test_split(X,y, test_size=0.2)
```

Menampilkan data test

```python
print(X_test)
```

Melatih model dengan *entropy* sebagai kriteria pengukuran proses splitting

```python
import sklearn.tree as tree
model1 = tree.DecisionTreeClassifier(criterion='entropy',max_depth=5)
model1.fit(X_train, y_train)
```

Uji coba dengan dataset test dan menampilkan hasilnya

```python
y_prediksi = model1.predict(X_test)
print(y_prediksi)
```

Menghitung akurasi model

```python
import sklearn.metrics as met
print(met.accuracy_score(y_test, y_prediksi))
```

Kita bisa mencoba untuk mengganti kriteria *splitting* dari *entropy* menjadi *gini index*

```python
import sklearn.tree as tree
model1 = tree.DecisionTreeClassifier(criterion='gini',max_depth=5)
model1.fit(X_train, y_train)

y_prediksi = model1.predict(X_test)
print(y_prediksi)

import sklearn.metrics as met
print(met.accuracy_score(y_test, y_prediksi))
```

Visualisasi

```python
import pydotplus as pp
labels = ['mesin','bangku','penggerak']
dot_data = tree.export_graphviz(model1, out_file=None, feature_names=labels, filled=True, rounded=True)
graph = pp.graph_from_dot_data(dot_data)
graph.write_png('decisiontree.png')
```
